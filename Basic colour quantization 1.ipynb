```python
%matplotlib notebook
import facets_improc as ip
import matplotlib.pyplot as plt
import numpy as np

img = ip.imread("Example Dithering 2.png")

# Colour quantization

# 1. Convert image to grayscale 
# 2. Convert image so it only uses black and white
# 3. Convert greyscale image so it uses 5 values
# 4. Convert original image so that each colour has 5 values 
# 5. ???

ip.imshow(img)
```

```python
# 1. 

#see Solomon and Breckon section 1.4.1.1
alpha = 0.2989  # Contribution of red
beta = 0.5870   # Contribution of green
gamma = 0.1140  # Contribution of blue

img_gs = alpha * img[:, :, 0] + beta * img[:, :, 1] + gamma * img[:, :, 2]

ip.imshow(img_gs)
```

```python
# 2. 

img_gs_round = img_gs
img_gs_round = 255*(img_gs_round>127)

ip.imshow(img_gs_round)
```

```python
# 3.

myList = [0, 64, 128, 192, 255]  #I want to round each pixel in image to nearest value in list

img_gs_round2 = img_gs

#print(min(myList, key=lambda x:abs(x-39)))
#from 'https://stackoverflow.com/questions/12141150/from-list-of-integers-get-number-closest-to-a-given-value'
#print(img_gs[464][653])

for y in range(0, len(img_gs)):
    for x in range(0, len(img_gs[0])):
        img_gs_round2[y,x] = min(myList, key=lambda a:abs(a-img_gs_round2[y,x]))
        
#img_gs_round2[][] = min(myList, key=lambda x:abs(x-img_gs))

ip.imshow(img_gs_round2)
```

```python
# 4.

myList = [0, 64, 128, 192, 255]

img_4 = img

for colour in range(0, 2):
    for y in range(0, len(img)):
        for x in range(0, len(img[0])):
            img_4[y,x,colour] = min(myList, key=lambda a:abs(a-img_4[y,x,colour]))

ip.imshow(img_4)
```
